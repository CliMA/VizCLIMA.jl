```julia
using Plots
using NCDatasets
using Statistics: mean
```

```julia
# load data
data_folder = "/central/scratch/jiahe/les_risingubble/netcdf/";
analysis_folder = "/central/scratch/jiahe/les_risingubble/analysis/";
```

```julia
# load data
auxdata = NCDataset(data_folder*"DryRisingBubble_DumpAux-2020-06-26T17.06.46.03.nc", "r");
statedata = NCDataset(data_folder*"DryRisingBubble_DumpState-2020-06-26T17.06.46.03.nc", "r");
```

```julia
t = statedata["time"][:];
x = statedata["x"][:];
y = statedata["y"][:];
z = statedata["z"][:];

xlen = size(x)[1]
ylen = size(y)[1]
zlen = size(z)[1]

ρ = statedata["ρ"];
ρ_ref = auxdata["ref_state.ρ"];

ρ_a = ρ[:] - ρ_ref[:];
b = -ρ_a./ρ[:];

w = statedata["ρu[3]"][:]./ρ[:];
```

```julia
nlev = 50

anim = @animate for it = 1:size(t)[1]
    # Plot the fields.
    xzplot = contourf(x, z, b[:, Int(floor(ylen/2)), :, it]', color=:balance,
        levels=[-0.01:0.0005:0.01;], linewidth=0)
    
    yzplot = contourf(x, z, ρ_a[:, Int(floor(ylen/2)), :, it]', color=:balance,
        levels=[-0.015:0.001:0.015;], linewidth=0)

    # Arrange the plots side-by-side.
    plot(xzplot, yzplot, layout=(1,2), size=(1000,500),
         title=["buoyancy x-z at "*string(it) "density y-z"])
end

mp4(anim, fps = 30)
```
